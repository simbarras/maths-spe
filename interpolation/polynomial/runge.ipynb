# Phénomène de Runge

![runge](docs/runge.png)

Plus il y a de points, plus les imprécisions sur le côté sont importantes.

```python
# import
import numpy as np # maths
import matplotlib.pyplot as plt # graph

def f(x): # function f
    return 1./(1.+25*x**2)

xx = np.linspace(-1, 1, 1000) # 1000 points between -1 and 1
yy = f(xx) # function values
```

```python
# Show the interpolation with 10
x = np.linspace(-1, 1, 11)
y = f(x)
a = np.polyfit(x, y, 10)
p = np.polyval(a, xx)
plt.plot(xx, p, 'r', xx, yy, 'b', x, y, 'o')
plt.show()
```

```python
# show the interpolation with 20
x = np.linspace(-1, 1, 21)
y = f(x)
a = np.polyfit(x, y, 20)
p = np.polyval(a, xx)
plt.plot(xx, p, 'r', xx, yy, 'b', x, y, 'o')
plt.show()
```

```python
# show the interpolation with 30
x = np.linspace(-1, 1, 31)
y = f(x)
a = np.polyfit(x, y, 30)
p = np.polyval(a, xx)
plt.plot(xx, p, 'r', xx, yy, 'b', x, y, 'o')
plt.show()
```

Chaque fois que l'on ajoute un point, l'oscillation augmente au extrémité de la courbe.

Essayons avec moins de points.

```python
# show the interpolation with 5
x = np.linspace(-1, 1, 6)
y = f(x)
a = np.polyfit(x, y, 5)
p = np.polyval(a, xx)
plt.plot(xx, p, 'r', xx, yy, 'b', x, y, 'o')
plt.show()
```

Nous voyons que nous sommes pas assez précis pour trouver une approximation de la courbe.

## La solution

La solution est de faire de l'interpolation linéaire par morceaux (spline). Cette solution consiste à relier les points de la courbe par des segments de droites. Ceci ressemble au dessin que l'on faisait dans les magazines disney.
